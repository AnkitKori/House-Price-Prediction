#### For categorical data replace null values with the most frequently used value of each feature.

```python
#replace categorical null values with most frequent values present in each column
categorical_data = df.select_dtypes(['object']).columns
df[categorical_data] = df[categorical_data].fillna(df[categorical_data].mode().iloc[0])
df[categorical_data].mode()
```

#### For numerical data replace null values with the mean value of each feature

```python
#replace numerical null values with the mean value of each column

numerical_data = df.select_dtypes(['float64','int64']).columns
df[numerical_data] = df[numerical_data].fillna(df[numerical_data].mean())
df[numerical_data].mean()
```

#### Model building

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into attributes and class
X = df.drop(['SalePrice'],axis=1)
y = df['SalePrice']
```

```python
#perform training and testing split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

#### Normalization

```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing dataabs
X_test_norm = norm.transform(X_test)
```
